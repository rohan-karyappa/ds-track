```python
def extract_episodes(txt):
  check = False
  data=''
  for i in txt:
    if i==')':
      check = False
      break
    if i=='(':
      check = True
    if check == True:
      if i=='(':
        continue
      data+=i
  return data
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('anime.csv')
print(df.loc[1]['Title'])
#Extract the number of episodes from the title
df['Episodes']=df['Title'].apply(extract_episodes)
df['Episodes']=df['Episodes'].str.replace(' eps','')
df['Episodes']=df['Episodes'].astype(int)
print(df)
```

```python
#MAKE A NEW COLUMN FOR TIMESTAMPS
def extract_timestamps(txt):
  check=False
  data=''
  for i in txt:
      if i==')':
        check=True
      if check==True:
        if i==')':
          continue
        data+=i
      if len(data)>18:
        break
  return data
df['Timestamps']=df['Title'].apply(extract_timestamps)
df.head()
```

```python
#ANIME WITH THE HIGHEST SCORE
highest_score = df[df['Score'] == df['Score'].max()]
print(highest_score)

#ANIME WITH THE MOST EPISODES
df[df['Episodes'] == df['Episodes'].max()]['Title']

#TOP 5 HIGHEST RATED
top_5 = df.nlargest(5, 'Score')[['Title','Episodes','Score']]
print(top_5)

#TOP 5 EPISODES
most_episodes = df.nlargest(5,'Episodes')[['Title','Episodes','Score']]
print(most_episodes)
```
