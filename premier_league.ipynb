### Loading the Libraries and Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv('IPL.csv')

df.head()

```

Q1. Which team won the most matches?

```python
match_wins = df['match_winner'].value_counts()
print(match_wins)
sns.barplot(y = match_wins.index,x = match_wins.values,palette='viridis')
plt.title("Match wins by team ")
```

2. Toss Decision Trends

```python
sns.countplot(x = df['toss_decision'],palette='rainbow')
plt.title("Toss Decision Trends ")
```

3. Toss Winner vs Match Winner

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count *100)/df.shape[0]
percentage.round(2)
```

4. How do teams win? (Runs vs Wickets)

```python
sns.countplot(x = df['won_by'])
plt.title("Won by")
```

###  Key Player Performances

Q1 - Most "Player of the Match" Awards

```python
count = df['player_of_the_match'].value_counts().nlargest(10)
count
sns.barplot(y=count.index,x=count.values,palette='rainbow')
```

Q2 - Top Scorers

```python
top_scorers = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(10)
top_scorers.plot(kind='barh')
```

Q3 - 10 Best Bowling Figures

```python
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wickets']=df['highest_wickets'].astype(int)
top_bowlers=df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
print(top_bowlers)
top_bowlers.plot(kind='barh')
```

###  Venue Analysis

Most Matches Played by Venue

```python
venue_counts = df['venue'].value_counts()
sns.barplot(y=venue_counts.index, x=venue_counts.values)
```
