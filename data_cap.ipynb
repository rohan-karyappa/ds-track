```python
import numpy as np
import pandas as pd
df = pd.read_csv('Countries.csv')
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
#country with the highest population
df[df['population'] == df['population'].max()]['country'] #target data frame, then write the condition
```

```python
#what is the capital of the country with highest population
df[df['population']==df['population'].max()]['capital_city']
```

```python
#country with the least population
df[df['population']==df['population'].min()]['country']
```

```python
#capital of the country with the least population
df[df['population']==df['population'].min()]['capital_city']
```

```python
#top seven countires with the highest democratic score
df.sort_values('democracy_score',ascending=False,inplace=True)
df[0:7]['country']
```

```python
#how many total regions are there
df['region'].value_counts().count()
```

```python
#how many countries lie in eastern european region and their names
print(df[df['region']=='Eastern Europe']['country'])
print(df[df['region']=='Eastern Europe']['country'].count())
```

```python
#political leader of the 2nd highest populated country
df[df['population']==df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
#countries where political leaders are unknown
df[df['political_leader'].isna()]['country']
```

```python
count = 0
def has_republic(txt):
  global count
  if 'republic' in txt.lower():
    count+=1
    return True
  return False
df[df['country_long'].apply(has_republic)]
```

```python
#which country in africa has the highest population
africa_df = df[df['continent']=='Africa']
africa_df[africa_df['population']==africa_df['population'].max()]
```

```python
#which country in africa has the highest population
df[df['continent']=='Africa'].nlargest(1,'population')
```
